#### Q1: Check Duplicate Records

```python
import pandas as pd
import numpy as np
df = pd.read_csv('cars - cars.csv')
df
df.duplicated().sum()
```

#### Q2: Drop Duplicate Records

```python
df = df.drop_duplicates()
df
```

#### Q3: Check missing or null values

```python
df.isnull()
```

#### Q4: Drop missing or null values

```python
df.dropna()
```

#### Q5: Show top 10 average MSRP (Price) per car brand (rounded figure)?

```python
df.groupby('Make')['MSRP'].mean().round().sort_values(ascending=False).head(10)
```

#### Q6: Show the average MSRP (price) per years (rounded figure)?

```python
df.groupby('Year')['MSRP'].mean().round()
```

#### Q7: Show average Engine HP (power) rounded figure and MPG (efficiency) rounded figure per year?

Hint: Average of (city mpg + highway MPG)

```python
df['Average_MPG'] = (df['city mpg'] + df['highway MPG']) / 2
df.groupby('Year')[['Engine HP', 'Average_MPG']].mean().round()
```

#### Q8: What percentage of cars are Compact, Midsize, or Large?

```python
(df['Vehicle Size'].isin(['Compact', 'Midsize', 'Large']).mean() * 100).round(2)
```

#### Q9: Which vehicle styles are the most sold (most common) in the dataset?

#### Q10: Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

#### Q11: Check if there are outliers in Engine HP show boxplot

#### Q13: Find upper bound, lower bound, number of outliers in Engine HP ?

```python

Q1 = df['Engine HP'].quantile(0.25)
Q3 = df['Engine HP'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Engine HP'] < lower_bound) | (df['Engine HP'] > upper_bound)]
lower_bound, upper_bound, outliers.shape[0]
```

#### Q14: Fill missing values in Number of Doors column with median value then check if values are filled ?

```python
df['Number of Doors'] = df['Number of Doors'].fillna(df['Number of Doors'].median())
df['Number of Doors'].isnull().sum()
```
